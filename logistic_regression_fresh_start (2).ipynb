```
import pandas as pd
from sklearn .linear_model import LogisticRegression
```

```
#LOAD THE DATASET
citizen=pd.read_csv("bankfull.csv")
citizen.head()
```

```
citizen.drop(["job","education","default"],inplace=True,axis = 1)
```

```
citizen.head()
```

```
citizen.shape
```

```
citizen=citizen.dropna()
citizen.shape
```

```
x = citizen.iloc[:,1:13]
y = citizen.iloc[:,-1]
```

```
x
```

```
y
```

```
#encoding

data_encoded=pd.get_dummies(x)
data_encoded
```

```
#label incoding for dependant veriable or reponse variable 
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```
labelencoder = LabelEncoder()
citizen1 = pd.DataFrame(labelencoder.fit_transform(y))
```

```
  citizen1
```

```
citizen2= data_encoded.join(citizen1)
citizen2
```

```
# selected the column and according to x and y (x = independant variable , y = dependent variable ( response variable ))
X = citizen2.iloc[:,0:32]
Y = citizen2.iloc[:,-1]
```

```
classifier = LogisticRegression()
classifier.fit(X,Y)
```

```
y_pred = classifier.predict(X)
```

```
y_pred_df = pd.DataFrame({'actual': Y,
                          # 'prob': classifier.predict_proba(X)[:,1],
                         'predicted': classifier.predict(X)})
```

```
y_pred_df[20:80]
```

```
y_pred_df['correct'] = y_pred_df['actual'] == y_pred_df['predicted']
y_pred_df['correct'].mean()
```

```
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y,y_pred)
print (confusion_matrix)
```

```
#Classification report
from sklearn.metrics import classification_report
print(classification_report(Y,y_pred))
```
